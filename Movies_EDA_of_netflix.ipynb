```python
def categorize_col(df,cols,labels):
  edges = [df[cols].describe()['min'],
           df[cols].describe()['25%'],
           df[cols].describe()['50%'],
           df[cols].describe()['75%'],
           df[cols].describe ()['max']]
  df[cols] = pd.cut(df[cols], edges,labels = labels ,  duplicates= 'drop')
  return df
```

```python
labels = ['not popular' , 'below average' , 'average', 'popular']
```

```python
# Convert 'Vote_Average' to numeric, coercing errors to NaN
df['Vote_Average'] = pd.to_numeric(df['Vote_Average'], errors='coerce')

# Drop rows with NaN in 'Vote_Average' after conversion
df.dropna(subset=['Vote_Average'], inplace=True)

def categorize_col(df,cols,labels):
  edges = [df[cols].describe()['min'],
           df[cols].describe()['25%'],
           df[cols].describe()['50%'],
           df[cols].describe()['75%'],
           df[cols].describe ()['max']]
  df[cols] = pd.cut(df[cols], edges,labels = labels ,  duplicates= 'drop')
  return df

categorize_col(df,'Vote_Average',labels)
```

```python
df['Genre'] = df['Genre'].str.split(', ')
df = df.explode('Genre').reset_index(drop=True)
```

Data Visualization

```python
sns.set_style('whitegrid')

sns.catplot(y ='Genre', data = df, kind = 'count',
            order = df['Genre'].value_counts().index,
            color = '#0000FF')
plt.title("genre distribution")
plt.show()
```

```python
sns.catplot(y = 'Vote_Average', data = df , kind = 'count',
            order = df['Vote_Average'].value_counts().index,
            color = '#0000FF')
plt.title("vote distribution")
plt.show()
```
